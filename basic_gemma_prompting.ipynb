```python
!pip install -q transformers accelerate bitsandbytes
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig

quantization_config = BitsAndBytesConfig(load_in_8bit=True)

tokenizer = AutoTokenizer.from_pretrained("google/gemma-2b-it")
model = AutoModelForCausalLM.from_pretrained("google/gemma-2b-it", quantization_config=quantization_config)




```

```python
promt = "Write me a poem about Machine Learning."
input_ids = tokenizer(promt, return_tensors="pt").to("cuda")

outputs = model.generate(**input_ids)
print(tokenizer.decode(outputs[0]))
```

```python
prompt = '''
<start_of_turn>user
You are a helpful 2nd-grade teacher. Help a 2nd grader to answer questions in a short and clear manner.
Explain why the sky is blue<end_of_turn>
<start_of_turn>model
'''
input_ids = tokenizer(prompt, return_tensors="pt").to("cuda")

outputs = model.generate(**input_ids, max_length=200)
print(tokenizer.decode(outputs[0]))
```

```python
prompt = '''
<bos><start_of_turn>user
Write a hello world program<end_of_turn>
<start_of_turn>model
'''
input_ids = tokenizer(prompt, return_tensors="pt").to("cuda")

outputs = model.generate(**input_ids, temperature=0.7, max_length=200, top_k = 50, top_p= 0.95)
print(tokenizer.decode(outputs[0]))
```

```python
prompt = '''
<bos><start_of_turn>user
Write a hello world program<end_of_turn>
<start_of_turn>model
'''
input_ids = tokenizer(prompt, return_tensors="pt").to("cuda")

outputs = model.generate(**input_ids, temperature=0.4, max_length=200, top_k = 50, top_p= 0.95, do_sample=True)


def clean_output(text):
  # Replace control codes and unnecessary tokens with empty strings
  cleaned_text = text.replace("<bos>", "").replace("<eos>", "").replace("<pad>", "")
  cleaned_text = cleaned_text.replace("<start_of_turn>", "").replace("<end_of_turn>", "")
  # You can add more replacements here based on Gemma's specific tokens
  return cleaned_text

decoded_text = tokenizer.decode(outputs[0])

# Limpiar la salida usando la función definida
cleaned_text = clean_output(decoded_text)

# Imprimir la salida limpia
print(cleaned_text)
```
